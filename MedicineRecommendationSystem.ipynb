```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
X=df.drop("prognosis",axis=1)
y=df["prognosis"]
```

```python
#LabelEncoding
le=LabelEncoder()
Y=le.fit_transform(y)
Y
```

```python
#TrainTestSplit
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
#Models
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB

from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
#ModelDictionary
models={
    "SVC":SVC(kernel="linear"),
    "RandomForest":RandomForestClassifier(n_estimators=100,random_state=42),
    "GradientBoosting":GradientBoostingClassifier(n_estimators=100,random_state=42),
    "KNN":KNeighborsClassifier(n_neighbors=5,),
    "MultinomialNB":MultinomialNB()
}

for model_name,model in models.items():
    model.fit(x_train,y_train)
    predictions=model.predict(x_test)
    accuracy=accuracy_score(y_test,predictions)
    cm = confusion_matrix(y_test,predictions)

    print(f"{model_name} accuracy : {accuracy}")
    print(f"{model_name} confusion matrix :")
    print(np.array2string(cm,separator = ","))
```

```python
#SinglePrediction
svc=SVC(kernel="linear")
svc.fit(x_train,y_train)
y_pred=svc.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
precautions=pd.read_csv("/content/drive/MyDrive/Datasets/precautions.csv")
workout=pd.read_csv("/content/drive/MyDrive/Datasets/workout.csv")
description=pd.read_csv("/content/drive/MyDrive/Datasets/description.csv")
medications=pd.read_csv("/content/drive/MyDrive/Datasets/medications.csv")
diets=pd.read_csv("/content/drive/MyDrive/Datasets/diets.csv")
```

```python
sym_dict = {col: idx for idx, col in enumerate(df.columns[:-1])}
diseases_list = {idx: disease for idx, disease in enumerate(le.classes_)}
```

```python
#HelperFunction
def helper (dis):
    desc = description[description['Disease'] == dis]['Description']
    desc = " ".join([w for w in desc])

    pre = precautions[precautions['Disease'] == dis][['Precaution_1', 'Precaution_2', 'Precaution_3', 'Precaution_4']]
    pre = [col for col in pre.values]

    med = medications[medications['Disease'] == dis]['Medication']
    med = [med for med in med.values]

    die = diets[diets['Disease'] == dis]['Diet']
    die = [die for die in die.values]

    wrkout = workout[workout['disease'] == dis]['workout']


    return desc, pre, med, die, wrkout


#WorkerFunction
def get_predicted_value(patient_symptoms):
  input_vector=np.zeros(len(sym_dict))

  for item in patient_symptoms:
    input_vector[sym_dict[item]]=1

  return diseases_list[svc.predict([input_vector])[0]]
```

```python
#Testing
symptoms=input("Enter Your Symptoms :")
user_sym=[s.strip() for s in symptoms.split(",")]
user_sym=[sym.strip("[]' ") for sym in user_sym]
predicted_disease=get_predicted_value(user_sym)
desc,pre,med,die,wrkout = helper(predicted_disease)

print("==============================Predicted Disease==============================")
print(predicted_disease)
print("==============================Description==============================")
print(desc)
print("==============================Precautions==============================")
n=1
for i in pre[0]:
  print(n," : ",i)
  n+=1
print("==============================Medicines==============================")
n=1
for i in med:
  print(n," : ",i)
  n+=1
print("==============================Diets==============================")
n=1
for i in wrkout:
  print(n," : ",i)
  n+=1
print("==============================Workout==============================")
n=1
for i in wrkout:
  print(n," : ",i)
  n+=1
```
